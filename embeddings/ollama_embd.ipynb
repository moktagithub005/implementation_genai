```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
##connect to ollama's embeddings model

embeding_model=OllamaEmbeddings(model="nomic-embed-text")

## example documents

documents=[
    "Alpha is the first character of the greek alphabet",
    "beta is the second letter of the greek alphabet"
]

# generate embeddings
embedding=embeding_model.embed_documents(documents)


```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

## step 1 initialize embedding model

embeding_model=OllamaEmbeddings(model="nomic-embed-text")

## write your doc

documents=[
    "the Eiffel tower located in paris",
    "the great wall of china can be seen from space",
    "mount everest is the highest mountain in the world",
    "python is the popular programming language for AI",
    "langchin helps you build apps with llms"
]

## step 3 store them with faiss

vector_store=FAISS.from_texts(documents,embeding_model)

## step 4 search 
query="what is the tallest peak on earth"

results=vector_store.similarity_search(query)

print(results)
```
